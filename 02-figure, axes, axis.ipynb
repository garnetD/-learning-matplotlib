## 三者关系

- figure: 相当于画板, 图的载体
- axes: 相当于画板上的画纸, 即一幅幅图. 由于axes是单词axis的复数, 因此可以理解为由坐标轴形成的区域为图
- axis: 具体的某个坐标轴(x, y等)

```python
import matplotlib.pyplot as plt
```

使用plt.figure()将会创建一个新的画板Figure, 创建画板之后, 就可以在画板上添加画纸Axes了, 一个Figure中可以添加多个Axes, 这就是我们所说的**子图**

常用的画图方法plot, bar, hist, imshow等方法, 是对Axes对象添加元素, 进行设置等

也可以用直接的set方法来完成

```python
fig = plt.figure(1)
print(fig)
ax1 = fig.add_subplot(111) # 注意这里是在Figure上添加的子图(Axes)
print(ax1)
ax1.set(xlim=[1, 2], ylim=[-1, 1], title="an example", xlabel="x axis", ylabel="y axis")
plt.show()
```

## 在Figure中添加子图Axes的方式

在**matplotlib**中, 添加子图有两种形式

#### fig.add_subplot() / plt.subplot()

传入一个三位的整数, 其中最高位表示子图总行数, 中间位表示子图总列数, 最低位表示当前子图是第几幅子图. 子图之间相互不重叠

add_subplot与subplot方法并无区别, 只不过

- add_subplot是面向对象的调用方法, 可以明显看到是调用了Figure实例的方法
- subplot是符合matlab使用习惯的方法

#### fig.add_axes()

add_axes方法其实是创建了多个轴形成的一个子图, 因此, 子图的位置将没有任何限制, 将会更灵活, 当然使用起来也会更麻烦

创建的时候传入一个**含有4个值的列表**, 每个值都在0到1之间, 代表的是这个Figure的比例

- 其中前两个值为坐标轴原点的位置, 即原点在Figure上的x, y坐标
- 后面两个是长度和高度的值

```python
fig = plt.figure()
ax1 = fig.add_subplot(121)
ax2 = fig.add_subplot(122)
print(ax1)
print(ax2)
plt.show()
```

```python
fig = plt.figure()
ax1 = fig.add_axes([0, 0, 1, 1])
ax2 = fig.add_axes([0.5, 0.3, 0.2, 0.6])
print(ax1)
print(ax2)
plt.show()
```

另外, 可以调用**plt.subplots(row_num, col_num)**方法, 获取一个Figure和Axes列表, 注意方法名字的最后有个's', 表示多个Axes

```python
fig, ax = plt.subplots(2, 2)
fig, ax
```

```python
fig, ax = plt.subplots(2, 2)
ax[1, 0].plot([1, 2, 3], c="g")
plt.show()
```

## 操作例子

常用的画图方法plot, bar, hist, imshow实际上都是对Axes进行操作, 从下面的例子可以看出

```python
fig = plt.figure()
ax1 = fig.add_subplot(311)
ax2 = fig.add_subplot(312)
ax3 = fig.add_subplot(313)
ax2.plot([1, 2, 3], c="r", ls="--")
ax1.set_xlim(10, 20)
ax3.imshow([[1, 2], [3, 4]])
plt.show()
```

## gca与gcf

两者都可以使用plt直接调用, 即plt.gca()和plt.gcf()

- gca()作用是: 获取当前Axes
- gcf()作用是: 获取当前Figure

首先, 按编号创建两个Figure. 由于后创建了fig2, 因此, 当前的Figure应该是fig2, 可以从下面的结果确定

```python
fig1 = plt.figure(1)
fig2 = plt.figure(2)

plt.gcf() is fig2
```
