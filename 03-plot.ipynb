```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
```

## 1. 三种使用形式

**plt.plot()** 是用来画直线图的, 对于传入的

有以下几种常用的调用方式:

- plt.plot(y, **kwargs)

    只传入y值, 即纵坐标的值, 横坐标x为默认的0, 1, 2, ...


- plt.plot(x, y, **kwargs)

    x值为横坐标, y值为纵坐标


- plt.plot(x1, y1, *args1, x2, y2, *args2, ...)
    
    - 传入多组(x, y, *args)对, 每一组代表一条线, 参数为这条线的设置
    - 由于每条线使用的参数的名称是一样的, 因此不能再使用字典形式传参, 直接在点数组对后紧跟参数值即可, 且将**所有的**参数放在一个字符串中, 函数内部会自动识别各个参数

```python
# 第一种情况: plt.plot(y, **kwargs)
plt.plot([2, 2, 1, 3], c="g")
```

```python
# 第二种情况: plt.plot(x, y, **kwargs)
plt.plot([1, 2, 5, 6], [1, 2, 3, 4], "g")
```

```python
# 第三种情况: plt.plot(x1, y1, args1, x2, y2, args2, ...)

t = np.arange(0, 5, 0.5)
plt.plot(
    t, t, "--b",
    t, t**2, "-*g",
    t, t**3, "^r"
)
```

```python
# 常用的颜色和线形状/点形状设置

t = np.arange(0, 5, 0.5)
plt.plot(
    t, t, 'b-',       # b蓝色, -实线
    t, t**1.1, 'g--', # g绿色, --虚线
    t, t**1.2, 'r-.', # r红色, -.虚点线
    t, t**1.3, 'y:',  # y黄色, :点线
    # 以上是线形状, 下面的都是点形状, 也可以在scatter(散点图)中使用, 在plot(线图)中使用时, 将会只画出点, 点之间的线不会被绘出
    t, t**1.4, 'co',  # c青色, o圆点
    t, t**1.5, 'm.',  # m品红, .细点
    t, t**1.6, 'k+',  # k黑色, +加号
    t, t**1.7, 'x',   # x乘号
    t, t**1.8, '*',   # *五角星
    t, t**1.9, 'd',   # 瘦菱形
    t, t**2.0, 'D--', # 方菱形, 注意, 这里可以同时设置线和点
)
```

## 2. 返回线对象

可以看到, plt.plot()方法返回元素为matplotlib.lines.Line2D对象的列表, **Line2D**即为一条线在程序中的实例. plot()方法画几条线, 就返回几个Line2D实例.

```python
t = np.arange(0, 5, 0.5)
line1 = plt.plot(t, t)
line2 = plt.plot(t, t, t, t**2)
print(len(line1), len(line2))
```

## 3. 设置线属性

一般在plot时, 直接指定线的各种属性

但也可以在画线之后, 调用plt.setp()方法, 对当前指定的一条线进行属性的设置, 需要传入这条线的实例对象

```python
x = np.linspace(-np.pi,np.pi)
y = np.sin(x)

lines = plt.plot(x, y) # 原本为蓝色实线, 线宽为1

# 使用键值对
plt.setp(lines, color='r', ls=':', linewidth=2.0)
'color', 'r', 'linestyle', '--', 'linewidth', 2,
plt.show()
```

还可以使用**matlab**格式, 使用字符串指定变量名, 后面紧跟参数值来进行设置

```python
x = np.linspace(-np.pi,np.pi)
y = np.sin(x)

lines = plt.plot(x, y) # 原本为蓝色实线, 线宽为1
plt.setp(lines, 'color', 'r', 'linestyle', '--', 'linewidth', 1, 'marker', '+', 'markersize', 10)

plt.show()
```
